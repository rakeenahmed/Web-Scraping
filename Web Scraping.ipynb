### Web Scraping Books Example

Our goal for this project is to grab the title of every book that has a minimum 2 stars rating.

To Do:
1. Scrape all books on the first page [Done]
2. Scrape only the books on the first page that have a 2 star rating [Done]
3. Do the same as above, but grab all books that have a minimum rating of 2 and above [Done]
4. Do the same as above, but scrape all pages of the website [Done]

```python
#requesting the webpage
response = requests.get('http://books.toscrape.com/')
```

```python
response.status_code
```

```python
#parsing the webpage with beautiful soup
soup = BeautifulSoup(response.content,'html.parser')
```

```python
#grabbing the first star rating
soup.select_one('.star-rating')
```

```python
#finding all star ratings
# only first two displayed
soup.find_all('p',class_ = ['star-rating Two', 'star-rating Three', 'star-rating Four', 'star-rating Five'])[0:2]
```

```python
#grabbing a single title
soup.select_one('.col-xs-6 h3').text
```

```python
#grabbing a single title, alternate method
soup.select_one('.product_pod h3 a').text
```

```python
#grabbing all titles
for item in soup.select('.col-xs-6 h3'):
    print(item.text)
```

Scrape all books on the first page that have a 2 star rating

```python
#grabbing only titles that have a 2 star rating
products = []
for item in soup.find_all('article',class_='product_pod'):
    if item.find('p',class_='star-rating Two'):
        products.append(item.h3.text)
    else:
        continue
products
```

```python
len(products)
```

Scrape all books on the first page that have a minimum rating of 2 and above

```python
#grabbing all titles that have a minimum 2 star rating and above
products = []
for item in soup.find_all('article',class_='product_pod'):
    if item.find('p',class_=['star-rating Two', 'star-rating Three', 'star-rating Four', 'star-rating Five']):
        products.append(item.h3.a['title'])
    else:
        continue
products
```

```python
len(products)
```

```python
#printing all titles in the products list
for item in products:
    print(item)
```

Scrape all pages in the website for books that have a minimum rating of 2 and above. Set it all up one cell as a script

```python
url = 'http://books.toscrape.com/catalogue/page-{x}.html'

products = [] #products list needs to be outside the for loop to prevent creating empty lists for each page

#the whole script has to be inside a giant for loop
#looping through the pages
for x in range(1,51):
    response = requests.get(url.format(x=x))    

    #creating the beautiful soup object
    soup = BeautifulSoup(response.content,'html.parser')

    #grabbing all titles that have a minimum 2 star rating and above
    for item in soup.find_all('article',class_='product_pod'):
        if item.find('p',class_=['star-rating Two', 'star-rating Three', 'star-rating Four', 'star-rating Five']):
            products.append(item.h3.a['title'])
        else:
            continue
    products
```

```python
len(products)
```

```python
books = pd.DataFrame(products,columns=['Title'])
books
```

```python
books.to_csv('books.csv')
```

### Wychwood BIA 

```python
response = requests.get('https://wychwoodheight.ca/bia-business-directory-all/')

soup = BeautifulSoup(response.content,'html.parser')

#writing the script to extract the details of each business in the directory
businesses = []
for item in soup.select('.cn-entry'):
    business = {}
    if item.select('.org.fn.notranslate') == []:
        pass
    else:
        business['name'] = item.select('.org.fn.notranslate')[0].text
    if item.select('.street-address') == []:
        pass
    else:
        business['address'] = item.select('.street-address')[0].text
    if item.select('.email.cn-email-address') == []:
        pass
    else:
        business['email'] = item.select('.email.cn-email-address')[0].text.split()[2]
    if item.select('.cn-contact-block') == []:
        pass
    else:
        business['contact'] = item.select('.cn-contact-block')[0].text.split()[1]
    if item.select('.tel.cn-phone-number') == []:
        pass
    else:
        business['phone'] = item.select('.tel.cn-phone-number')[0].text
    if item.select('.link.cn-link.website') == []:
        pass
    else:
        business['website'] = item.select('.link.cn-link.website')[0].text

    businesses.append(business)
    

businesses = pd.DataFrame(businesses)

#saving the data as csv
businesses.to_csv('businesses.csv')
```

### Yahoo Finance

```python
response = requests.get('https://ca.finance.yahoo.com/crypto')
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup.select_one('table tr td a').text
```

```python
coins = []

for item in soup.select('table tr'):
    coin = {}
    if item.select('a') == []:
        continue
    else:
        coin['symbol'] = item.select('a')[0].text
        coin['name'] = item.select('td')[1].text
        coin['price'] = item.select('td')[2].text
        coin['market cap'] = item.select('td')[5].text
    coins.append(coin)
```

```python
coins = pd.DataFrame(coins)
```

### ATN Mega Store

```python
url = 'https://atnmegastore.ca/products/books/page/{x}/'

response = requests.get(url.format(x=x))

books = []

for x in range(1,175):
    response = requests.get(url.format(x=x))

    soup = BeautifulSoup(response.content,'html.parser')

    for item in soup.select('.product .product-details'):
        book = {}
        book['title'] = item.select('h3 a')[0].text.strip()
        book['price'] = item.select('.price')[0].text
        book['url'] = item.select('h3 a')[0]['href']
        book['image url'] = item.parent.parent.select('img')[0]['src']
        books.append(book)
```

```python
len(books)
```

```python
books = pd.DataFrame(books)
```

```python
books.head()
```

```python
books.to_csv('books.csv')
```

### Steam Top Sellers

```python
response = requests.get('https://store.steampowered.com/search/?filter=topsellers')
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup.select_one('#search_resultsRows a .title').text
```

```python
games = []
for item in soup.select('#search_resultsRows a'):
    game = {}
    game['title'] = item.select('#search_resultsRows a .title')[0].text
    game['relese_date'] = item.select('.col.search_released ')[0].text
    if item.select('.discount_final_price') == []:
        game['price'] = 'free to play'
    else:
        game['price'] = item.select('.discount_final_price')[0].text
    games.append(game)
```

### Greektown BIA

```python
categories = ['restaurants-bars-cafes','entertainment-media','convenience-store','business-professional-services','financial-services-institutions','personal-care-services','health-wellness']

url = 'https://greektowntoronto.com/business-directory/{x}'

businesses = []

for x in categories:
    response = requests.get(url.format(x=x),headers={'User-Agent':'Mozilla/5.0'})
    soup = BeautifulSoup(response.content,'html.parser')
    print(response.status_code)
    
    for item in soup.select('.businessListing'):
        business = {}
        business['name'] = item.select('.businessTitle')[0].text
        business['address'] = item.select('.businessAddress')[0].text
        business['phone'] = item.select('.businessPhone')[0].text
        if item.select('.businessWebsite') == []:
            pass
        else:
            business['website'] = item.select('.businessWebsite')[0].text
        business['category'] = x
        businesses.append(business)
```

```python
businesses = pd.DataFrame(businesses)
```

```python
businesses.shape
```

```python
businesses.head()
```

```python
businesses.to_csv('greektown_businesses.csv')
```
